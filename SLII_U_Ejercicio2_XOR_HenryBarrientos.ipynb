```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
```

## AND

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 0, 0, 1])

colours = ListedColormap(['r', 'b'])
classes = ["0", "1"]
plot = plt.scatter(X[:,0],X[:, 1],c=list(Y),cmap=colours)

plt.legend(handles=plot.legend_elements()[0],labels=classes)
plt.grid(True)
plt.show()
```

```python
weight1 = 1
weight2 = 1
bias = -1.5

outputs = X.copy()
linear_combination = (weight1*X[:,0] + weight2*X[:,1] + bias).reshape((4,1))
outputs = np.hstack(( outputs, linear_combination ))
outputs = np.hstack(( outputs, (outputs[:,2]>=0).astype(int).reshape((4,1)) ))   # 3:activation
outputs = np.hstack(( outputs, Y.reshape((4,1)) ))   # 4:real value

is_correct_col = outputs[:,3]==outputs[:,4]
output_frame = pd.DataFrame(outputs, columns=['Input 1', '  Input 2', '  Linear Combination', '  Activation Output', "Real value"])
output_frame['  Is Correct'] = is_correct_col.reshape((4,1))
num_wrong = len([ val for val in is_correct_col if not val ])
if (num_wrong == 0):
    print('Nice!  You got it all correct.\n')
else:
    print('You got {} wrong.  Keep trying!\n'.format(num_wrong))
print(output_frame.to_string(index=False))

line_x = np.linspace(0,1.3)
line_y = (-line_x*weight1 - bias)/weight2
sns.scatterplot(data = output_frame,x="Input 1",y="  Input 2",hue="Real value",style="  Activation Output")
plt.plot(line_x,line_y)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## XOR con perceptrones

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0,1,1,0])

colours = ListedColormap(['r','b'])
classes = ["0","1"]
plot = plt.scatter(X[:,0],X[:, 1],c=list(Y),cmap=colours)

plt.legend(handles=plot.legend_elements()[0],labels=classes)
plt.grid(True)
plt.show()
```

```python
def activacion_heaviside(z):
    return int(z >= 0)


def perceptron_AND(X1,X2):
    weight1 = 1
    weight2 = 1
    bias = -1.5
    z = weight1*X1 + weight2*X2  + bias
    h = activacion_heaviside(z)
    return h


def perceptron_OR(X1,X2):
    weight1 = 2
    weight2 = 2
    bias = -1.5
    z = weight1*X1 + weight2*X2  + bias
    h = activacion_heaviside(z)
    return h


def perceptron_NOT(X1,X2):
    weight1 = 0
    weight2 = -10
    bias =  5
    z = weight1*X1 + weight2*X2  + bias
    h = activacion_heaviside(z)
    return h


def perceptron_IDENTIDAD(X1,X2):
    return (X1,X2)
```

```python
X1 = X[0,0]
X2 = X[0,1]

r_and = perceptron_AND(X1, X2)
r_or = perceptron_OR(X1, X2)

r_not = perceptron_NOT(r_and, 0)

y_prediccion = perceptron_AND(r_not, r_or)
print("y_prediccion:",  y_prediccion)
```
